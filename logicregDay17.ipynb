```python
import statsmodels
import pandas as pd #data analysis and data manipulation
import numpy as np #numerical computations
import matplotlib.pyplot as plt  #for data visualization
```

```python
%matplotlib inline
```

```python
df=pd.read_csv('datafiles/claimants.csv')
df
```

```python
df.shape
```

```python
#removing casenum
df1=df.drop(['CASENUM'],axis=1,inplace=False)
df1.head()
```

```python
x=df1.iloc[:,1:]
x.head()
```

```python
y=df1.iloc[:,1]
y.head()
```

```python
# filling the missing value with most occuring value    
df1.iloc[:,:] = df1.iloc[:,:].apply(lambda x:x.fillna(x.value_counts().index[1]))
#claimants3.SEATBELT = claimants1.SEATBELT.fillna(claimants1.SEATBELT.value_counts().index[0])
df1.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression()
reg.fit(x_train,y_train)
```

```python
pred=reg.predict(x_test)
pred
```

```python
np.mean(y_test==pred)
```
