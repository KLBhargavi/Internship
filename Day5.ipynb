## Problem Statement:
Print number of matches won and loss(as two columns). Consider as win when a team wins Toss  and Match . print for each and every team


```python
import pandas as pd
import numpy as np
def wintoss(filename):
    df=pd.read_csv(filename)
    l1=[]
    d={}
    winlist=[]
    losslist=[]
    for i in df.values:
        l1.append(i[9])
    uniq=np.unique(l1)
    for j in uniq:
        win=0
        loss=0
        for i in df.values:
            if j==i[9] and j==i[10]:
                win=win+1
            elif j==i[2] or j==i[3]:
                loss=loss+1
        winlist.append(win)
        losslist.append(loss)
        
    d={'Teams':uniq,
       'Wins': winlist,
       'loses':losslist}
    dff=pd.DataFrame(d)
    return dff
#         # return df
wintoss('datafiles/Match.csv')
```
